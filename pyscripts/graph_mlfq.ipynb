```python
import csv
x = []
y1 = []
y2 = []
y3 = []
y4 = []
with open('data.csv', 'r') as csv_file:
    reader = csv.reader(csv_file)
    diff = 0
    for row in reader:
        x.append(int(row[0]) - diff) 
        diff = x[0]
        y1.append(int(row[5]))
        y2.append(int(row[6]))
        y3.append(int(row[7]))
        y4.append(int(row[8]))
```

```python
x[0] = 0
```

```python
import matplotlib.pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 10, 5

plt.xlabel ('Time')
plt.ylabel ('Queue Number')
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.plot(x, y4)
plt.legend(["Pure CPU", "Pure IO", "IO then CPU", "CPU then IO"], fontsize='small')
plt.savefig("graph.jpg")
plt.show()
```
